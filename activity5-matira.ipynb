```python
## IMPORING LIBRARIES

import numpy as np
import matplotlib.pyplot as plt
```

```python
## DEFINING VARIABLES

dt = 0.001 # step size
t = np.arange(0,1,dt)
fa = 50
fb = 20
fc = 130
aa = 5
ab = 10
ac =3
```

```python
## GENERATING SINE WAVES
a = aa * np.sin(2*np.pi*fa*t)
b = ab * np.sin(2*np.pi*fb*t)
c = ac * np.sin(2*np.pi*fc*t)

```

```python
## GENERATING NOISE
noise = np.random.normal(0,4,len(a))
```

```python
## COMPOUNDING SINE WAVES
comp = a + b + c + noise
```

```python
def plotSineWave(values, color, title):
    plt.figure(figsize=(15,7))
    plt.plot(t, values, color = color)
    plt.title(title)
    plt.ylabel("amplitude")
    plt.xlabel("time (s)")
    plt.show()
```

```python
## PLOTTING SINE WAVES INDIVIDUALLY

plotSineWave(a, "blue", "Sine Wave a")
plotSineWave(b, "red", "Sine Wave b")
plotSineWave(c, "green", "Sine Wave c")
plotSineWave(noise, "gray", "Noise")

```

```python
## PLOTTING SINE WAVES TOGETHER

plotSineWave(comp, "orange", "Compounde Sine Wave")
```

## PART 2

```python
## GENERATING SINE WAVE 
amp = 5
f = 10
wave = amp * np.sin(2*np.pi*f*t)
```

```python
plotSineWave(wave, "purple", "Raw Sine Wave")
```

```python
## GENERATING SQAURE WAVE
squareWave = []
for value in wave:
    if value >= -1 and value <= 1:
        squareWave.append(value)
    elif value >= 1:
        squareWave.append(1)
    elif value <= -1 :
        squareWave.append(-1)
```

```python
plotSineWave(squareWave, "green", "Sqaure Wave")
```

```python
## GENERATING TRIANGULAR WAVE
triangularWave = []
for n,value in enumerate(squareWave):
    if n == 0: triangularWave.append(value); continue
    triangularWave.append(triangularWave[n-1] + value)
```

```python
plotSineWave(triangularWave, "green", "Triangular Wave")
```
