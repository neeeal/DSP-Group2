## Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
## Generating random numbers
array = np.random.randint(low=0, high=255, size=(150,150))
```

```python
array[:3]
```

```python
## Displaying random numbers
plt.title("Randomly generated image")
plt.imshow(array, cmap='gray')
```

```python
## Generating kernel
kernel = [[1, -1, -1],
          [1, -1, -1,],
          [1, 1, 1]]
kernel=np.array(kernel)
kernel.shape
```

```python
## Displaying kernel
plt.title("Kernel")
plt.imshow(kernel, cmap='gray')
```

## Part 2

```python
## Algorithm for convolution
temp=np.zeros(shape=(148,148))
while True:
    try:
        for row in range(temp.shape[0]-2):
            for col in range(temp.shape[1]-2):
                temp[row,col] = (array[row:row+3,col:col+3]*kernel).sum()
        break
    except:
        break    
```

```python
## Original array 1st step 
array[0:3,0:3]
```

```python
## multipled array 1st step
array[0:3,0:3]*kernel
```

```python
## new pixel of array after convolution
(array[0:3,0:3]*kernel).sum()
```

```python
## Shape of images
print('Shape of old image', array.shape)
print('Shape of new image', temp.shape)
```

```python
## Displaying convolved image
fig, ax = plt.subplots(1,2)
ax[0].set_title("Original image")
ax[0].imshow(array, cmap='gray')
ax[1].set_title("Convolved image")
ax[1].imshow(temp, cmap='gray')
```

## Part 3

```python
## convolved array
temp[0:3]
```

```python
## normalizing array 
norm_temp = temp.copy()
for row in range(temp.shape[0]):
    for col in range(temp.shape[1]):
        norm_temp[row, col] = (temp[row,col]-np.min(temp))/(np.max(temp)-np.min(temp))*255
```

```python
## Displaying convolved image
fig, ax = plt.subplots(1,3, figsize=(12,7))
ax[0].set_title("Original image")
ax[0].imshow(array, cmap='gray')
ax[1].set_title("Convolved image")
ax[1].imshow(temp, cmap='gray')
ax[2].set_title("Nromalized image")
ax[2].imshow(norm_temp, cmap='gray')
```
